## Bubble Sort

```python
def bubble_sort(arr:List[int]):
    for i in range(len(arr)):
        print(f"Iter. {i} --------------------------------------------")
        print(f"\tcomparing {arr[i]} with the elements in {arr[i+1:]}")
        for j in range(len(arr)-i-1):
            if arr[j] > arr[j+1]:
                # swap
                print(f"{arr} === after swapping {arr[j]} and {arr[j+1]} ==> ", end='')
                arr[j], arr[j+1] = arr[j+1], arr[j]
                print(f"{arr[0:i+1]}{arr[i+1:]}")
    return arr
```

## Selection Sort

## Insertion Sort

```python
def insertion_sort(arr:List[int]):
    for i in range(1, len(arr)):
        print(f"Iter. {i} --------------------------------------------")
        print(f"\tcomparing {arr[i]} with the elements in {arr[i+1:]}")
        for j in range(i, 0, -1):
            if arr[j-1] > arr[j]:
                print(f"{arr} === after swapping {arr[j-1]} and {arr[j]} ==> ", end='')
                arr[j-1], arr[j] = arr[j], arr[j-1]
                print(f"{arr[0:i+1]}{arr[i+1:]}")
    return arr
```

## Merge Sort (Non-inplace)

```python
# 두 개의 정렬된 리스트를 병합하는 함수
def merge(left, right):
    len_left, len_right = len(left), len(right)  # 각 리스트의 길이를 저장
    i_left, i_right = 0, 0  # 각 리스트의 현재 인덱스를 추적하기 위한 변수
    l = []  # 병합된 결과를 저장할 리스트
    print(f"{l} {left[i_left:], right[i_right:]}")  # 현재 상태 출력
    
    # 두 리스트 모두에서 아직 처리되지 않은 항목이 있는 동안 반복
    while (i_left < len_left) and (i_right < len_right):
        # left의 현재 항목이 right의 현재 항목보다 작으면 left의 항목을 결과에 추가
        if left[i_left] < right[i_right]:
            l.append(left[i_left])
            i_left += 1
        # 그렇지 않으면 right의 항목을 결과에 추가
        else:
            l.append(right[i_right])
            i_right += 1
        print(f"{l} {left[i_left:], right[i_right:]}")  # 현재 상태 출력
    
    # left에 남아있는 항목들을 결과에 추가
    while i_left < len_left:
        l.append(left[i_left])
        i_left += 1
        print(f"{l} {left[i_left:], right[i_right:]}")  # 현재 상태 출력
        
    # right에 남아있는 항목들을 결과에 추가
    while i_right < len_right:
        l.append(right[i_right])
        i_right += 1
        print(f"{l} {left[i_left:], right[i_right:]}")  # 현재 상태 출력
    
    return l  # 병합된 결과 반환


# 주어진 리스트를 병합 정렬하는 함수
def merge_sort(arr:List[int], n):
    # n: 현재 재귀 호출의 깊이에 따라 출력 메시지 앞에 들여쓰기(indentation)를 추가하는 역할
    # 리스트의 길이가 0 또는 1이면 이미 정렬된 것으로 간주하고 그대로 반환
    if (len(arr) == 0) or (len(arr) == 1):
        return arr
    mid = int(len(arr) / 2)  # 리스트를 둘로 나누기 위한 중간 인덱스 계산
    print(f"{'  '*n}Dividing {arr} into {arr[0:mid]}{arr[mid:]}")  # 현재 상태 출력
    left = merge_sort(arr[:mid], n + 1)  # 리스트의 왼쪽 절반을 재귀적으로 정렬
    print(f"{'  '*n}After sorting {arr[:mid]}, {left}")  # 정렬된 상태 출력
    right = merge_sort(arr[mid:], n + 1)  # 리스트의 오른쪽 절반을 재귀적으로 정렬
    print(f"{'  '*n}After sorting {arr[mid:]}, {right}")  # 정렬된 상태 출력
    r = merge(left, right)  # 두 개의 정렬된 리스트를 병합
    print(f"{'  '*n}After sorting {left} and {right}, {r}")  # 병합된 결과 출력
    return arr
```

## Quick Sort (Non-inplace)

```python
def partition(arr:List[int], low:int, high:int):
    p = arr[low]  # 피벗을 선택합니다.
    small, large = [], []  # 피벗보다 작은 원소들과 큰 원소들을 저장할 리스트를 초기화합니다.
    print(f"Input: [{p}]{arr[low+1:]}")  # 현재 상태를 출력합니다.
    
    for i in arr[low+1:high+1]:  # 피벗을 제외한 나머지 원소들을 순회합니다.
        if i < p:
            small.append(i)  # 피벗보다 작은 원소는 small 리스트에 추가합니다.
        else:
            large.append(i)  # 피벗보다 큰 원소는 large 리스트에 추가합니다.
        print(f"[{p}] {small} {large}")  # 현재 상태를 출력합니다.
    
    small.append(p)  # 피벗을 small 리스트의 마지막에 추가합니다.
    print(f"{small} {large}")  # 현재 상태를 출력합니다.
    
    idx = low
    for i in small:  # small 리스트의 원소들을 arr에 복사합니다.
        arr[idx] = i
        idx += 1
    for i in large:  # large 리스트의 원소들을 arr에 복사합니다.
        arr[idx] = i
        idx += 1
    print(arr)  # 현재 상태를 출력합니다.
    
    return low + len(small) - 1  # 피벗의 위치를 반환합니다.


def quick_sort(arr:List[int]):
    def qs(arr:List[int], low:int, high:int, n:int):
        print(f"{'===='*n} Sorting {arr[low:high+1]}")  # 현재 정렬 대상 범위를 출력합니다.
        if low < high:
            pivot_idx = partition(arr, low, high)  # 피벗을 기준으로 분할합니다.
            print(f"{'===='*n} After partitioning: {arr[low:pivot_idx]}[{arr[pivot_idx]}]{arr[pivot_idx+1:high+1]}")  # 분할 후의 상태를 출력합니다.
            qs(arr, low, pivot_idx-1, n+1)  # 피벗보다 작은 부분을 재귀적으로 정렬합니다.
            qs(arr, pivot_idx+1, high, n+1)  # 피벗보다 큰 부분을 재귀적으로 정렬합니다.
            print(f"{'===='*n} After partitioning: {arr[low:pivot_idx]}[{arr[pivot_idx]}]{arr[pivot_idx+1:high+1]}")  # 재귀 호출 후의 상태를 출력합니다.
        return arr
    qs(arr, 0, len(arr)-1, 0)  # 전체 배열에 대해 퀵 정렬을 시작합니다.
    return arr
```

## Counting Sort

```python
from typing import List

def counting_sort(arr:List[int], m:int):
    # arr의 원소들은 [0, m] 범위 내에 있다고 가정합니다.
    
    # m+1 크기의 새로운 리스트 c를 초기화하고, 모든 원소를 0으로 설정합니다.
    # 다른 방법으로는 c = [0 for i in range(m)]를 사용할 수도 있습니다.
    c = [0 for _ in range(m+1)]
    
    # arr의 각 원소 i에 대해 c[i]를 1씩 증가시킵니다.
    # 이렇게 하면 c[i]는 arr에서 i가 나타나는 횟수를 나타냅니다.
    for i in arr:
        c[i] += 1
    print(c)
    
    # c[j]는 j보다 작거나 같은 키의 개수를 저장합니다.
    # 따라서 키 j는 최종 정렬된 배열의 인덱스 c[j-1], ..., c[j]-1에 저장되어야 합니다.
    for j in range(1, m+1):
        c[j] = c[j-1] + c[j]
    print(c)
    
    n = len(arr)
    # 정렬된 결과를 저장할 임시 리스트 t를 초기화합니다.
    t = [0 for _ in range(n)]
    
    # arr의 원소들을 거꾸로 순회하면서 정렬된 위치에 값을 저장합니다.
    for i in range(n-1, -1, -1):
        j = arr[i]
        t[c[j]-1] = j
        c[j] = c[j] - 1
    
    # 최종 정렬된 결과를 원래의 arr에 복사합니다.
    for i in range(n):
        arr[i] = t[i]
    return arr
```

## Radix Sort

```python
from typing import List

def radix_sort(arr:List[int], d:int):
    # 가장 낮은 자릿수부터 시작하여 정렬을 수행합니다.
    # d: 정렬할 숫자의 최대 자릿수를 나타냅니다.
    # radix_sort는 숫자 외의 데이터 형식에는 적용할 수 없습니다.
    for r in range(d):
        # 사실상 counting_sort와 동일한 로직을 사용합니다.
        # 0부터 9까지의 숫자를 카운트하기 위한 리스트 c를 초기화합니다.
        c = [0 for _ in range(10)]
        
        # 현재 정렬 대상인 자릿수를 결정합니다.
        m = d - r - 1
        for i in arr:
            # i의 m번째 자릿수를 key로 사용합니다.
            key = int(str(i)[m])
            c[key] += 1
        
        # c[j]는 j보다 작거나 같은 키의 개수를 저장합니다.
        for j in range(1, 10):
            c[j] = c[j-1] + c[j]
        
        n = len(arr)
        # 정렬된 결과를 저장할 임시 리스트 t를 초기화합니다.
        t = [0 for _ in range(n)]
        
        # arr의 원소들을 거꾸로 순회하면서 정렬된 위치에 값을 저장합니다.
        for i in range(n-1, -1, -1):
            key = int(str(arr[i])[m])
            t[c[key]-1] = arr[i]
            c[key] = c[key] - 1
        
        # 최종 정렬된 결과를 원래의 arr에 복사합니다.
        for i in range(n):
            arr[i] = t[i]
    return arr
```

# P2 Search 구현 및 성능 분석 보고서 작성

수업 시간에 배운 `linear_search` 및 `binary_search`를 `python`으로 구현하라. 그 후 입력 리스트의 크기를 `1`, `10`, `100`, `1000`, `10000`, `100000`으로 주어서 각 search 알고리즘을 실행하고 실행시간을 측정하여 실행시간을 비교할 수 있는 하나의 그래프를 도시하라. 

그래프의 X축은 입력의 크기, Y축은 실행시간(micro-seconds) 을 밑이 10인 log 스케일로 나타낸다. 입력은 random number를 생성하여 만들고 각각 다른 입력으로 5번 실행하여 실행시간의 평균을 취한다. 아래에 대하여 실행시간을 비교할 수 있는 각각 하나의 그래프가 나와야 한다. 

1. random number로 생성된 입력이 이미 sorting이 된 경우 
2. random number로 생성된 입력이 sorting이 되지 않았을 경우 

도시한 search algorithm 들의 실행시간에 대한 그래프를 보고 search 알고리즘의 big O time complexity와 실제 실행시간 간의 관계를 논하라. 

## Linear Search

```python
def linear_search(arr, target) -> bool:
    for item in arr:
        if item == target:
            return True
    return False


def linear_search_last_idx(arr, target) -> int:
    # target이 array에 있는 경우, target과 같은 item 가운데 가장 마지막에 있는 item의 index를 return
    # array에 없는 경우 -1
    idx = -1
    for i, item in enumerate(arr):
        if item == target:
            idx = i
    return idx 

def linear_search_first_idx(arr, target):
    # target이 array에 있는 경우, target과 같은 item 가운데 가장 첫 번째에 있는 item의 index를 return
    # array에 없는 경우 -1
    for i, item in enumerate(arr):
        if item == target:
            return i
    return -1
```

```python
## recursion을 쓰는 경우
def binary_search_bool(arr, target) -> bool:
    def binary_search_helper(arr, low, high, target):
        if low > high:
            return False 
        mid = (low + high) // 2
        if arr[mid] == target:
            return True
        if arr[mid] > target:
            return binary_search_helper(arr, low, mid-1, target)
        if arr[mid] < target:
            return binary_search_helper(arr, mid+1, high, target)
    return binary_search_helper(arr, 0, len(arr)-1, target) != -1


def binary_search_idx(arr, target) -> int:
    def binary_search_helper(arr, low, high, target):
        if low > high:
            return -1 
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] > target:
            mid = binary_search_helper(arr, low, mid-1, target)
        if arr[mid] < target:
            mid = binary_search_helper(arr, mid+1, high, target)
        return mid
    return binary_search_helper(arr, 0, len(arr)-1, target)
```

```python
## iterative
def binary_search_iter_idx(arr, target) -> int:
    low, high = 0, len(arr) - 1
    while (low < high):
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        if arr[mid] < target:
            low = mid + 1
        if arr[mid] > target:
            high = mid - 1
    return -1


def binary_search_iter_bool(arr, target) -> bool:
    low, high = 0, len(arr) - 1
    while (low < high):
        mid = (low + high) // 2
        if arr[mid] == target:
            return True
        if arr[mid] < target:
            low = mid + 1
        if arr[mid] > target:
            high = mid - 1
    return False
```
