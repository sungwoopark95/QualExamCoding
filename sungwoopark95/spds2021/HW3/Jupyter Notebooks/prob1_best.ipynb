```python
# I have copied the class code
def bubble_sort(l):
    def swap(i,j):
        # first save the first element
        t = l[i]
        # save the second element to first 
        l[i] = l[j]
        # save the first element to where the second element was
        l[j] = t
        return
    n = len(l) - 1
    for i in range(n,0,-1):
        for j in range(i):
            if (l[j] >= l[j+1]):
                swap(j,j+1)
    return l

def insertion_sort(l):
    n = len(l)
    for j in range(1,n):
        key = l[j]
        i = j-1
        while ((i>=0) and (key < l[i])):
            l[i+1] = l[i]
            i-=1
        l[i+1] = key
    return l

def selection_sort(l):
    def swap(i,j):
        t=l[i]
        l[i] = l[j]
        l[j] = t
        return
    n = len(l)
    for i in range(0, n-1):
        s = i
        swap_flag = False
        for j in range(i+1, n):
            if (l[s] >= l[j]):
                s = j
                swap_flag = True
        swap(i,s) if swap_flag else None  
        return l

def merge_sort(left, right):
    len_left, len_right = len(left), len(right)
    i_left, i_right = 0,0
    l=[]
    while ((i_left < len_left) and (i_right < len_right)):
        if (left[i_left] < right[i_right]):
            l.append(left[i_left])
            i_left +=1
        else:
            l.append(right[i_right])
            i_right +=1
    while (i_left < len_left):
        l.append(left[i_left])
        i_left +=1
    while (i_right < len_right):
        l.append(right[i_right])
        i_right +=1
    return l

def partition(l, low, high):
    def swap(i,j):
        t = l[i]
        l[i] = l[j]
        l[j] = t
        return
    
    p = l[low]
    m = low
    for j in range(low + 1, high + 1):
        if (l[j] < p):
            m+=1
            swap(j,m)
    swap(low, m)
    return m

def quick_sort(l):
    def qs(l, low, high, n):
        if (low < high):
            pivot_idx = partition(l, low, high)
            qs(l, low, pivot_idx - 1, n+1)
            qs(l, pivot_idx + 1, high, n + 1)
        return l
    qs(l, 0 , len(l) - 1, 0)
    return l


def counting_sort(l,m):
    c = [0] * (m + 1)
    for i in l:
        c[i] += 1
    
    for j in range(1, m + 1):
        c[j] = c[j-1] + c[j]
    
    n = len(l)
    t = [0] * n
    for i in range(n-1,-1,-1):
        j = l[i]
        t[c[j] - 1] = j
        c[j] = c[j] - 1
    for i in range(0,n):
        l[i] = t[i]
    return l

def radix_sort(l,d):
    for r in range(0,d):
        c = [0] * 10
        m = d - r - 1
        for i in l:
            key = int(str(i)[m])
            c[key] += 1
        for j in range(1,10):
            c[j] = c[j-1] + c[j]
        n = len(l)
        t = [0] * n
        for i in range(n-1,-1,-1):
            key = int(str(l[i])[m])
            t[c[key] - 1] = l[i]
            c[key] = c[key] - 1
        for i in range(0,n):
            l[i] = t[i]
        return l
```

```python
import random
from time import perf_counter
import numpy as np
import pandas as pd
```

```python
# best case
b1 = []
i1 = []
s1 = []
m1 = []
q1 = []
c1 = []
r1 = []
```

- Bubble Sort
    - Best  = sorted
    - Worst = reverse-sorted
- Insertion Sort
    - Best  = sorted
    - Worst = reverse-sorted
- Selection Sort
    - Best = Sorted
    - Worst = First is largest and the rest is in order
- Merge Sort
    - Best = Sorted
    - Worst = 
- Quick Sort =
    - Best = When Partition process always picks the middle element as pivot
    - Worst = When Partition process always picks greatest or smallest element as pivot
        - When the N elements are already in ascending order
- Counting Sort
    - Best = when all elements are same
    - Worst = when data is skewed and range is large
- Radix Sort
    - Best = All elements have the same number of digits
    - Worst = Only one element which has significantly large number of digits

```python
length = [1, 10, 100, 1000, 10000, 100000]
```

### Best Case

#### Bubble, Insert, Select
- Sorted

```python
for i in length:
    print(i)
    a = []
    # create list of numbers
    for j in range(i):
        a.append(random.randint(0,999))
    # Sort array of random integer
    a.sort()
    
    t1 = perf_counter()
    x = bubble_sort(a)
    t2 = perf_counter()
    elapsed = t2 - t1
    b1.append(elapsed)
    print('Bubble')
    
    t1 = perf_counter()
    x = insertion_sort(a)
    t2 = perf_counter()
    elapsed = t2 - t1
    i1.append(elapsed)
    print('Insert')
    
    t1 = perf_counter()
    x = selection_sort(a)
    t2 = perf_counter()
    elapsed = t2 - t1
    s1.append(elapsed)
    print('Select')
```

#### Merge
- Sorted 

```python
for i in length:
    print(i)
    a = []
    if i == 1:
        left = [random.randint(0,999)]
        right = []
    else:
        # create list of numbers
        for j in range(i):
            a.append(random.randint(0,999))
        # Sort array of random integer
        a.sort()
        left = a[:(len(a)//2)]
        right = a[(len(a)//2):]


    t1 = perf_counter()
    x = merge_sort(left, right)
    t2 = perf_counter()
    elapsed = t2 - t1
    m1.append(elapsed)
```

#### Quick
- When Partition process always picks the middle element as pivot
    - When all elements are same according to wikipedia
        - Found out this sometimes leads to error
    - So the idea was to have it so that it would be
        - [number, smaller, bigger, smaller, bigger, ...]

```python
for i in length:
    print(i)
    a = []
    # create list of numbers
    for j in range(i):
        if j == 0:
            a.append(random.randint(0,998))
        # odd slots, we want bigger once
        elif j % 2 == 0:
            a.append(random.randint(a[j-1],999))
        # even slots, we want smaller number
        else:
            a.append(random.randint(0,a[j-1]))

    t1 = perf_counter()
    x = quick_sort(a)
    t2 = perf_counter()
    elapsed = t2 - t1
    q1.append(elapsed)
```

#### Counting
    - When all elements are same

```python
for i in length:
    print(i)
    a = [random.randint(0,999)] * i
    t1 = perf_counter()
    x = counting_sort(a, 1000)
    t2 = perf_counter()
    elapsed = t2 - t1
    c1.append(elapsed)
```

#### Radix
- All elements have the same number of digits

```python
for i in length:
    print(i)
    a = [random.randint(0,999)] * i
    t1 = perf_counter()
    x = radix_sort(a,len(str(a[0])))
    t2 = perf_counter()
    elapsed = t2 - t1
    r1.append(elapsed)
```

```python
df = pd.DataFrame(list(zip(length, b1, i1,s1,m1,q1,c1,r1)),
              columns=['Length','Bubble', 'Insertion','Select','Merge','Quick','Count','Radix'])
```
