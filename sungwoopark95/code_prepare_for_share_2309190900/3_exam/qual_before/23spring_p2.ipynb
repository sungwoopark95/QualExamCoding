```python

class Node :
    def __init__(self, data, next) :
        self.data = data
        self.next = next

def print_list(s: Node) :
    if s==None :
        return None
        # or return "[]"
    print('[', end='')
    head = s
    while head :
        print(head.data, end='')
        head = head.next
        if head != None :
            print(', ', end='')
    print(']')

def palindrome(s:Node) :
    '''return 1 if palindrome else 0'''
    if s==None : return 1
    length=0
    curr=s
    val_list=[]
    while curr!=None :
        length+=1
        val_list.append(curr.data)
        curr=curr.next
    # print(f"length is {length}")
    # print(f"val_list is {val_list}")
    
    rear_pointer=length//2
    front_pointer=rear_pointer-1
    if length%2==1 : rear_pointer+=1

   
    while (front_pointer>=0) :
        # print(f"front_pointer is {front_pointer}\trear_pointer is {rear_pointer}")
        if val_list[front_pointer]!=val_list[rear_pointer] : 
            return 0
        front_pointer-=1
        rear_pointer+=1
    return 1

def sub_list(s:Node, t:Node) :
    '''is t in s? Yes:1 No:0'''
    if t==None : return 1
    if s==None : return 0

    ptr_t=t
    ptr_s=s

    while ptr_s is not None :
        if ptr_s.data==ptr_t.data :
            #check_whether_same : return 1 
            ptr_ss=ptr_s
            while (ptr_t is not None) and (ptr_ss is not None) and (ptr_t.data==ptr_ss.data):
                ptr_t=ptr_t.next
                ptr_ss=ptr_ss.next
            if ptr_t==None : 
                return 1

        ptr_t=t
        ptr_s=ptr_s.next
    return 0

def max_palindrome(s:Node) :
    ### print max_palindrome in linked list
    
    if s==None :
        print([])

    all_palindrome=[]

    ptr_start=s
    ptr_end=s
    while ptr_start!=None:
        ptr_end=ptr_start  
        while ptr_end!=None :
            tmp_ptr=ptr_end.next #hold
            # print(f"tmp_ptr.data={tmp_ptr.data}")
            ptr_end.next=None #cut temporarily
            if palindrome(ptr_start)==1 : 
                all_palindrome.append((ptr_start, ptr_end))
                tmp_ptr2=ptr_start
                print("Appended list : ", end="")
                while tmp_ptr2!=None:
                    print(f"{tmp_ptr2.data} ", end="")
                    tmp_ptr2=tmp_ptr2.next
                print("")
            ptr_end.next=tmp_ptr #원상복구
            # print(f"tmp_ptr.data={tmp_ptr.data}")
            ptr_end=ptr_end.next
            # print(f"here. ptr_end.data={ptr_end.data}")
        ptr_start=ptr_start.next
    for (start, end) in all_palindrome :
        flag=0
        for (super_start, super_end) in all_palindrome :
            if (start!=super_start or end!=super_end)\
                and sub_list(super_start, start)==1 \
                and sub_list(end, super_end)==1 : #contain
                flag=1
                break
        if flag==0 :
            tmp_ptr=end.next
            end.next=None
            print_list(start)
            end.next=tmp_ptr

    '''
    `max_palindrome` 함수를 작성하는 데는 여러 가지 접근 방법이 있습니다. 
    여기서는 `palindrome` 함수와 `sub_list` 함수를 활용하는 방법에 대해 설명하겠습니다.

1. **모든 부분 리스트 찾기**: 
먼저, 주어진 연결 리스트의 모든 부분 리스트를 찾아야 합니다. 
이를 위해, 두 개의 포인터를 사용할 수 있습니다. 
하나의 포인터는 시작 노드를 가리키고, 다른 하나는 끝 노드를 가리킵니다. 
시작 포인터를 연결 리스트의 처음부터 끝까지 이동시키면서, 
각 위치에서 끝 포인터를 시작 포인터부터 연결 리스트의 끝까지 이동시킵니다. 
이렇게 하면 모든 가능한 부분 리스트를 찾을 수 있습니다.

2. **회문 확인**: 각 부분 리스트에 대해 `palindrome` 함수를 호출하여 
해당 부분 리스트가 회문인지 확인합니다. 만약 회문이라면, 그 길이를 확인하고 이전에 찾은 가장 긴 회문의 길이와 비교합니다.

3. **최대 회문 업데이트**: 만약 새로 찾은 회문의 길이가 이전에 찾은 가장 긴 회문의 길이보다 크다면, 최대 회문을 업데이트합니다.

4. **결과 반환**: 모든 부분 리스트를 확인한 후, 가장 긴 회문을 반환합니다.

이 접근 방법은 모든 부분 리스트를 확인하므로 시간 복잡도는 O(n^3)입니다 (n은 연결 리스트의 길이). 따라서, 이 방법은 연결 리스트의 길이가 크지 않은 경우에만 효율적입니다. 만약 연결 리스트의 길이가 매우 크다면, 다른 접근 방법을 고려해야 할 수 있습니다.
    '''
```

```python
# Test 1: Palindrome list
node1 = Node('a', None)
node2 = Node('b', node1)
node3 = Node('a', node2)
print(palindrome(node3))  # Output: 1

# Test 2: Non-palindrome list
node1 = Node('a', None)
node2 = Node('b', node1)
node3 = Node('c', node2)
print(palindrome(node3))  # Output: 0

# Test 3: Single character list
node1 = Node('a', None)
print(palindrome(node1))  # Output: 1

# Test 4: Empty list
print(palindrome(None))  # Output: 1

# Test 5: Palindrome list with even length
node1 = Node('a', None)
node2 = Node('b', node1)
node3 = Node('b', node2)
node4 = Node('a', node3)
print(palindrome(node4))  # Output: 1
```

```python
# Test 1: t is a sublist of s
s1 = Node('a', Node('b', Node('c', None)))
t1 = Node('b', Node('c', None))
print(sub_list(s1, t1))  # Output: 1

# Test 2: t is not a sublist of s
s2 = Node('a', Node('b', Node('c', None)))
t2 = Node('b', Node('d', None))
print(sub_list(s2, t2))  # Output: 0

# Test 3: t is the same as s
s3 = Node('a', Node('b', Node('c', None)))
t3 = Node('a', Node('b', Node('c', None)))
print(sub_list(s3, t3))  # Output: 1

# Test 4: t is an empty list
s4 = Node('a', Node('b', Node('c', None)))
t4 = None
print(sub_list(s4, t4))  # Output: 1

# Test 5: s is an empty list
s5 = None
t5 = Node('a', None)
print(sub_list(s5, t5))  # Output: 0
```
