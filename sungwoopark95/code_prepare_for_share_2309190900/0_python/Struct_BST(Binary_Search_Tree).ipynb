```python
class Node():
    def __init__(self, x:int) :
        self.val = x
        self.left = None
        self.right = None
        
class BST():
    def __init__(self) :
        self.root=None
        
    def search(self, x:int) :
        return self.search_help(self.root, x)
    def search_help(self, cur_node:Node, x:int) :
        if cur_node==None :
            print(' \"No such valued node. \" ')
            return None
        if x == cur_node.val :
            return cur_node
        elif x < cur_node.val :
            return self.search_help(cur_node.left, x)
        else : # x > cur_node.val
            return self.search_help(cur_node.right, x)
        
    def insert(self, x:int) :
        self.root = self.insert_help(self.root, x)
    def insert_help(self, cur_node:Node, x:int) :
        try :
            if cur_node==None :
                return Node(x)
            if x < cur_node.val :
                cur_node.left=self.insert_help(cur_node.left, x)
            elif x > cur_node.val :
                cur_node.right=self.insert_help(cur_node.right, x)
            else : #x==cur_node.val
                print(f"The value ({x}) already exists in the BST.")
            return cur_node # 마지막에는 반드시 자기 자신을 회신하도록 해야함!!!!!!
        except Exception as e :
            print(f"An error occurred in insert_help function : {e}")
            
    def delete(self, x:int) :
        self.root = self.delete_help(self.root, x)
    def delete_help(self, cur_node, x) :
        if cur_node == None :
            print(f"No such valued node ({x}).")
            return None
        if x<cur_node.val :
            cur_node.left = self.delete_help(cur_node.left, x)
        elif x>cur_node.val :
            cur_node.right = self.delete_help(cur_node.right, x)
        else : #x==cur_node.val
            if cur_node.left==None and cur_node.right==None :
                return None
            elif cur_node.left==None:
                return cur_node.right
            elif cur_node.right==None:
                return cur_node.left
            else : #cur_node.left!=None and cur_node.right!=None
                min_val = self.find_min(cur_node.right)
                print(f"MIN_VAL is {min_val}.")
                cur_node.val=min_val
                cur_node.right = self.delete_help(cur_node.right, min_val)
        return cur_node #주의주의! 마지막에 반드시 자신을 회신하도록!
    def find_min(self, cur_node):
        '''
        cur_node가 None이면 에러남
        '''
        curr = cur_node 
        while curr.left!=None :
            curr=curr.left
        return curr.val

    def print_tree(self):
        if self.root is None:
            print("BST is empty")
        else:
            self._print_tree(self.root, 0)

    def _print_tree(self, cur_node, level):
        if cur_node is not None:
            self._print_tree(cur_node.right, level + 1)
            print("  " * level + str(cur_node.val))
            self._print_tree(cur_node.left, level + 1)
    
#     def print_tree(self):
#         self.print_tree_help(self.root, "")
#     def print_tree_help(self, cur_node:Node, indent:str):
#         if cur_node is not None:
#             self.print_tree_help(cur_node.right, indent + "   ")
#             print(indent + str(cur_node.val))
# #             self.print_tree_help(cur_node.left, indent + "   ")

#     def print_tree(self):
#         lines, _, _, _ = self.print_tree_help(self.root)
#         for line in lines:
#             print(line)
#     def print_tree_help(self, node):
#         if node is None:
#             return [], 0, 0, 0
#         label = str(node.val)
#         left_lines, left_pos, left_width, left_height = self.print_tree_help(node.left)
#         right_lines, right_pos, right_width, right_height = self.print_tree_help(node.right)
#         middle = max(right_pos + left_width - left_pos + 1, len(label), 2)
#         pos = left_pos + middle // 2
#         width = left_pos + middle + right_width - right_pos
#         height = max(left_height, right_height) + 2
#         label_len = len(label)
#         label_left = (middle - label_len) // 2
#         label_right = middle - label_len - label_left
#         lines = [' ' * width for _ in range(height)]
#         if node.left is not None:
#             lines[1] = (' ' * left_pos + '/' + ' ' * (middle-2) +
#                          '\\' + ' ' * right_width)
#             for i in range(left_height):
#                 lines[i+3] = (' ' * left_pos + left_lines[i] +
#                                ' ' * (width - left_pos - len(left_lines[i])))
#         if node.right is not None:
#             lines[1] = (' ' * left_pos + '/' + ' ' * (middle-2) +
#                          '\\' + ' ' * right_width)
#             for i in range(right_height):
#                 lines[i+3] = (' ' * (pos + 2) + right_lines[i] +
#                                ' ' * (width - pos - 2 - len(right_lines[i])))
#         lines[2] = (' ' * pos + label +
#                      ' ' * (width - pos - label_len))
#         return lines, pos, width, height
```
