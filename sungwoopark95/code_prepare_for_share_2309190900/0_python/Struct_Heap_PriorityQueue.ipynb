# Max-Heap
## value가 큰 값이 parent node에 오도록

### 0-indexed list에서 
 - i번째 node의 child는 2i+1, 2i+2
 - i번째 node의 parent는 (i-1)//2

```python
#heap sort - 최대 heap 으로-


# Ftn making heap structure /  n : last index
def buildHeap(A : list, n : int)->None :
    for i in range( (n-1)//2,-1,-1) :
        heapify(A, i, n)
    return


# Ftn making heap structure under A[k] / k:corresponding node / n : last index
def heapify(A:list, k : int, n : int) -> None : 

    # A[k]와 하위 노드(A[2*k+1], A[2*k+2]) 간의 관계 정리. (단, 하위 노드가 n 이내임을 확인하기)
    if 2*k+1 > n : #하위 노드가 없으면 return
        return
    elif 2*k+ 1 == n : #하위 노드가 1개이면 비교 대상으로 선정
        larger, l_index = A[2*k+1], 2*k+1
    else : # 하위 노드가 2개이면 비교 대상은 큰 것으로 선정
        if A[2*k+1] >= A[2*k+2] :
            larger, l_index = A[2*k+1], 2*k+1
        else : 
            larger, l_index = A[2*k+2], 2*k+2

            
    if larger > A[k] : # A[k] 노드가 하위 노드보다 값이 작면 
        A[k], A[l_index] = A[l_index], A[k] # 서로 바꾸고,
        heapify(A, l_index, n) #하위 노드도 heapify 해준다
    return

def heapSort(A:list, n: int)-> list :
    buildHeap(A, n)
    print("after building Heap:", A)
    for i in range(n,0,-1):
        A[0], A[i] =A[i], A[0]
        heapify(A,0, i-1)
        print("A :",A, "when iteration is done with i :", i)
        
def heap_max(A:list) :
    if len(A) ==0 : 
        print("heap underflow")
    else :
        return A[0]

def heap_extract_max(A:list) :
    val = heap_max(A)
    A[0] = A[len(A)-1]
    A.pop()#remove last element
    heapify(A, 0, len(A)-1)
    return val
```

```python
import heapq

class PriorityQueue:
    def __init__(self):
        self._queue = []
        self._entry_finder = {} #key : task, value : entry=[priority, self._counter, task]
        self._REMOVED = '<removed-task>'
        self._counter = 0

    def add_task(self, task, priority=0):
        ## remove the origin one
        if task in self._entry_finder:
            self.remove_task(task)
        entry = [priority, self._counter, task] #counter는 priority가 같을 경우에 순서를 비교하기 위한 용도
        self._counter += 1
        self._entry_finder[task] = entry
        heapq.heappush(self._queue, entry)

    def remove_task(self, task):
        entry = self._entry_finder.pop(task)
        entry[-1] = self._REMOVED #heap 내에 있는 entry도 수정이 됨(alias이기 때문)

    def pop_task(self):
        while self._queue:
            priority, count, task = heapq.heappop(self._queue)
            if task is not self._REMOVED:
                del self._entry_finder[task]
                return task
        raise KeyError('pop from an empty priority queue')

# 사용 예제
pq = PriorityQueue()
pq.add_task('Apple', 3)
pq.add_task('Banana', 1)
pq.add_task('Cherry', 2)

# 'Banana'의 우선순위를 변경합니다.
pq.add_task('Banana', 4)

while True:
    try:
        print(pq.pop_task())
    except KeyError:
        break
```
