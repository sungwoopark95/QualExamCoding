# 1. Convolution algorithm 구현하기

- image: m x m np array
- pattern: n x n np array
- stride도 감안해야 함. Padding, Dilation 등은 없음
- m < n 이거나, stride와 m,n이 정확히 맞아 떨어지지 않으면 Error 출력    

```python
def conv(image,pattern,stride = 1):
    beg = 0
    end = pattern.shape[0]
    res = []
    
    m = image.shape[0]
    n = pattern.shape[0]
    
    if m < n:
        print("Error: m < n")
        return None
    
    if (m - n) % stride > 0:
        print("Incompatible patterns")
        return None
    
    for i in range(0,m-n+1,stride):
        k = []
        for j in range(0,m-n+1,stride):
            if i + end > m or j + end > m: continue
            
            k.append(np.sum(image[beg+i : end+i, beg+j:end+j] * (pattern)))
        res.append(k)

    return np.array(res)

```

## 기홍 답안

```python
def conv(image, pattern, stride=1):
    # image: a 2d numpy array of floats, with size (m, m) for some integer m.
    # pattern: a 2d numpy array of floats, with size(n, n) for some integer n.
    # stride: the number of pixels shift over the input.
    # return: a 2d numpy of convoluted activation on image with pattern if valid. None otherwise.
    m = len(image)
    n = len(pattern)

    if m < n:
        print("Error: m < n")
        return None
    if (m - n) % stride != 0:
        print("Incompatible params")
        return None

    result_width = (m - n) // stride + 1
    result_height = (m - n) // stride + 1
    result_array = np.zeros((result_width, result_height))

    for start_col in range(0, result_width):
        for start_row in range(0, result_height):  
            curr_col = start_col * stride
            curr_row = start_row * stride
            curr_output = image[curr_row:curr_row + n, curr_col: curr_col + n] * pattern
            curr_sum = np.sum(curr_output)
            result_array[start_row][start_col] = curr_sum
    
    return result_array
```

# 2. Tree가 주어졌을 때 가장 긴 Round Trip Length를 구하는 문제

```python
class BTNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
def roundTripLongestRouteOfBT(root: BTNode):
    def helper(node):
        if node.left is None and node.right is None:
            return 0, 0            
        if node.left is None:
            roundLen, singleLen = helper(node.right)
            return max(roundLen, singleLen + 1), singleLen + 1
        if node.right is None:
            roundLen, singleLen = helper(node.left)
            return max(roundLen, singleLen + 1), singleLen + 1
        l_roundLen, l_singleLen = helper(node.left)
        r_roundLen, r_singleLen = helper(node.right)
        roundLen = max(l_roundLen, r_roundLen)
        return max(roundLen, l_singleLen + r_singleLen + 2), max(l_singleLen, r_singleLen) + 1
    
    return 2 * max(helper(root))
```

## 기홍 답안

```python
# Definition for binary tree node
class BTNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = 0
        self.left = left
        self.right = right

class DiameterTracker:
	def __init__(self):
		self.diameter = 0

def roundTripLongestRouteOfBT(root: BTNode) -> int:
    diameter_tracker = DiameterTracker()
    getLongestPath(root, diameter_tracker)
    return diameter_tracker.diameter * 2
	
def getLongestPath(node, diameter_tracker):
	if node is None:
	    return 0
	left_longest_path = getLongestPath(node.left, diameter_tracker)
	right_longest_path = getLongestPath(node.right, diameter_tracker)
	diameter_tracker.diameter = max(diameter_tracker.diameter, left_longest_path + right_longest_path)
	return max(left_longest_path, right_longest_path) + 1
```

```python
# Test Cases
root_node = BTNode(4)
root_node.left, root_node.right = BTNode(2), BTNode(6)
root_node.left.right = BTNode(3)
root_node.right.right = BTNode(7)

root2 = BTNode(9)
root2.right = BTNode(6)
root2.right.left, root2.right.right = BTNode(3), BTNode(7)
root2.right.left.left = BTNode(2)
root2.right.right.left = BTNode(10)

print(roundTripLongestRouteOfBT(root_node))
print(roundTripLongestRouteOfBT(root2))
```
